```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_excel("/content/drive/MyDrive/Raw_Data_FDA.xlsx")


```

```python
# Check first few rows
print(df.head())
```

```python
# Get summary statistics
print(df.info())
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Drop rows with missing values
df.dropna(inplace=True)
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Check for duplicate values
print(df.duplicated().sum())
```

```python
# drop duplicate values
df.drop_duplicates(inplace=True)
print(df.duplicated().sum())
```

```python
#Distribution of Income
plt.figure(figsize=(8, 5))
sns.histplot(df['Annual Income'], bins=30, kde=True, color='blue')
plt.title("Income Level Distribution")
plt.xlabel("Income Level")
plt.ylabel("Count")
plt.show()
```

```python
# Bar Plot - Laptop Purchase
plt.figure(figsize=(10, 5))
sns.countplot(x='Occupation', data=df, order=df['Occupation'].value_counts().index, palette='coolwarm')
plt.xticks(rotation=45)
plt.title("Laptop Purchases by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Count")
plt.show()
```

```python
print(df.columns)
```

```python
#Income Level by Market
plt.figure(figsize=(10, 5))
sns.boxplot(x='Market ', y='Annual Income', data=df, palette='viridis')
plt.title("Income Level Across Markets")
plt.xlabel("Market ")
plt.ylabel("Income Level")
plt.show()
```

```python
# Scatter Plot of Income vs Laptop Price
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Annual Income', y='Laptop Price', data=df, hue='Market ', palette='coolwarm', alpha=0.7)
plt.title("Income vs. Laptop Price")
plt.xlabel("Income Level")
plt.ylabel("Laptop Price")
plt.show()
```
